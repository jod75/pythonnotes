# 10. Python notes - Fun with trigonometric functions

Throughout this notebook, we make use of [matplotLib](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.html) and [numpy](https://numpy.org/) libraries.

```python
import matplotlib.pyplot as plt # For ploting
import numpy as np # to work with numerical data efficiently
```

```python
fs = 100.0 # sample rate 
f = 2 # the frequency of the signal
```

```python
x = np.arange(fs) # the points on the x axis for plotting

# compute the value (amplitude) of the sin wave at the for each sample
y = np.sin(2 * np.pi * f * (x/fs)) 
```

```python
# this instruction can only be used with IPython Notbook. 
%matplotlib inline

plt.title("Voltage sample over time")
plt.stem(x,y, 'r', )
plt.xlabel('sample(n)')
plt.ylabel('voltage(V)')
plt.plot(x,y)
```

```python
x = np.arange(-np.pi, np.pi, 0.1)
y = np.cos(x)
```

```python
plt.title('Cosine wave')
plt.xlabel('x')
plt.ylabel('Amplitude = cos(x)')
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(-np.pi, np.pi, 0.1)
y = np.tan(x)
```

```python
plt.title('Tangent wave')
plt.xlabel('theta')
plt.ylabel('Amplitude = tan(theta)')
plt.grid(True, axis='x')
plt.axhline(y=0, color='k')
plt.plot(x,y)
plt.show()
```
